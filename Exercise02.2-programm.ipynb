```python
import json
import requests
import pandas as pd
response = requests.get("https://raw.githubusercontent.com/dis-data-modeling-2020/slides/master/movies.json")
datastore = response.json()
```

```python
genres = {}
for year in datastore:
    genres.update({year : {'Comedy' : 0, 'Drama' : 0, 'Action' : 0, 'Biography' : 0, 'Crime' : 0, 'Music' : 0, 'Western' : 0, 'Romance' : 0, 'Mystery' : 0, 'Animation' : 0, 'Short' : 0, 'Family' : 0, 'Adventure' : 0, 'Game-Show' : 0, 'War' : 0, 'History' : 0, 'Documentary' : 0, 'Film-Noir' : 0, 'Musical' : 0, 'Fantasy': 0, 'Thriller': 0, 'Talk-Show':0, 'Sci-Fi':0, 'Horror':0, 'Reality-TV':0, 'Sport':0, 'Ohne Genre': 0}})
    for title in datastore[year]:
        for genre in datastore[year][title]['genre']:
            if(genre == 'Comedy'):   
                genres[year]['Comedy'] += 1
            elif(genre == 'Drama'):
                genres[year]['Drama'] += 1 
            elif(genre == 'Action'):
                genres[year]['Action'] += 1 
            elif(genre == 'Biography'):
                genres[year]['Biography'] += 1 
            elif(genre == 'Crime'):
                genres[year]['Crime'] += 1 
            elif(genre == 'Music'):
                genres[year]['Music'] += 1 
            elif(genre == 'Western'):
                genres[year]['Western'] += 1 
            elif(genre == 'Romance'):
                genres[year]['Romance'] += 1 
            elif(genre == 'Mystery'):
                genres[year]['Mystery'] += 1 
            elif(genre == 'Animation'):
                genres[year]['Animation'] += 1 
            elif(genre == 'Family'):
                genres[year]['Family'] += 1 
            elif(genre == 'Short'):
                genres[year]['Short'] += 1 
            elif(genre == 'Adventure'):
                genres[year]['Adventure'] += 1 
            elif(genre == 'Game-Show'):
                genres[year]['Game-Show'] += 1 
            elif(genre == 'War'):
                genres[year]['War'] += 1
            elif(genre == 'History'):
                genres[year]['History'] += 1 
            elif(genre == 'Documentary'):
                genres[year]['Documentary'] += 1 
            elif(genre == 'Film-Noir'):
                genres[year]['Film-Noir'] += 1 
            elif(genre == 'Musical'):
                genres[year]['Musical'] += 1
            elif(genre == 'Fantasy'):
                genres[year]['Fantasy'] += 1
            elif(genre == 'Thriller'):
                genres[year]['Thriller'] += 1
            elif(genre == 'Talk-Show'):
                genres[year]['Talk-Show'] += 1
            elif(genre == 'Sci-Fi'):
                genres[year]['Sci-Fi'] += 1
            elif(genre == 'Horror'):
                genres[year]['Horror'] += 1
            elif(genre == 'Reality-TV'):
                genres[year]['Reality-TV'] += 1 
            elif(genre == 'Sport'):
                genres[year]['Sport'] += 1
            else:
                genres[year]['Ohne Genre'] +=1
                
genresJson = json.dumps(genres)
```

```python
import csv
df = pd.read_json(genresJson)
df.to_csv(r'Exercise02.2-Filmgenres.csv')
newCsv = pd.read_csv('Exercise02.2-Filmgenres.csv')
genresDataframe = pd.DataFrame(newCsv)
genresDataframe
```
